````python
from kneed import KneeLocator
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def find_optimal_k(data, max_k=10):
    wcss = []

    for k in range(1, max_k+1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)

    # Automatically detect elbow point
    kl = KneeLocator(range(1, max_k+1), wcss, curve="convex", direction="decreasing")

    print(f"Optimal k (Elbow Point): {kl.elbow}")
    return kl.elbow

    # Plot the elbow method
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_k+1), wcss, marker='o', linestyle='-', color='b', label="WCSS")
    plt.axvline(kl.elbow, linestyle="--", color="red", label=f"Elbow at k={kl.elbow}")
    plt.xlabel("Number of Clusters (k)")
    plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
    plt.title("Elbow Method for Optimal k")
    plt.legend()
    plt.grid(True)
    plt.show()

# Usage:
optimal_clusters = find_optimal_k(coordinates, max_k=10)
````

##### My own K-Means clustering implementation

````python
import numpy as np 

def my_kmeans(X, k, max_iters=10000, tol=1e-4, random_state=42):
    """
    A simple K-means clustering implementation.

    Parameters:
    -----------
    X : numpy.ndarray, shape (n_samples, n_features)
        Your data to be clustered.
    k : int
        The number of clusters.
    max_iters : int
        Maximum number of iterations to run the K-means algorithm.
    tol : float
        Tolerance threshold for declaring convergence. If the shift of all centroids
        is below this value, we stop early.
    random_state : int
        Seed for reproducibility.

    Returns:
    --------
    labels : numpy.ndarray, shape (n_samples,)
        Cluster index for each sample.
    centroids : numpy.ndarray, shape (k, n_features)
        Final cluster centroid positions.
    """
    np.random.seed(random_state)

    # 1. Initialize centroids by randomly picking k distinct points from X
    random_indices = np.random.choice(X.shape[0], k, replace=False)
    centroids = X[random_indices, :]

    for iteration in range(max_iters):
        # 2. Assign each sample to the closest centroid
        #    - Compute distances from each sample to each centroid
        distances = np.linalg.norm(X[:, None] - centroids[None, :], axis=2)
        #    - Find the cluster index (centroid) with minimal distance
        labels = np.argmin(distances, axis=1)

        # 3. Compute new centroids based on the mean of samples in each cluster
        new_centroids = np.array([X[labels == c].mean(axis=0) for c in range(k)])

        # 4. Check how far centroids have moved
        centroid_shifts = np.linalg.norm(new_centroids - centroids, axis=1)
        if np.all(centroid_shifts < tol):
            print(f"Converged after {iteration} iterations.")
            break

        centroids = new_centroids

    return labels, np.array(centroids)
    
labels, centroids = my_kmeans(coordinates, k=optimal_clusters)

print("Cluster labels:", labels)
print("Centroids:\n", centroids)
xyCoord['Manual Clustering'] = labels
xyCoord.head()
````
